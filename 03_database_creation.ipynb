```python
import pandas as pd
from sqlalchemy import create_engine

races = pd.read_csv("f1_races_2000_2026.csv")
results = pd.read_csv("f1_race_results_2000_2026.csv")
qualifying = pd.read_csv("f1_qualifying_2000_2026.csv")
drivers = pd.read_csv("f1_driver_standings_2000_2026.csv")
constructors = pd.read_csv("f1_constructor_standings_2000_2026.csv")

engine = create_engine("sqlite:///raceiq.db")
```

```python
races.to_sql("races", engine, if_exists="replace", index=False)

results.to_sql("results", engine, if_exists="replace", index=False)

qualifying.to_sql("qualifying", engine, if_exists="replace", index=False)

drivers.to_sql("driver_standings", engine, if_exists="replace", index=False)

constructors.to_sql("constructor_standings", engine, if_exists="replace", index=False)

print("Tables Created Successfully!")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(
        text("SELECT name FROM sqlite_master WHERE type='table';")
    )

    for row in result:
        print(row)
```
