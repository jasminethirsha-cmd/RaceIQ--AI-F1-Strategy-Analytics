```python
import pandas as pd

races = pd.read_csv("f1_races_2000_2026.csv")

results = pd.read_csv("f1_race_results_2000_2026.csv")

qualifying = pd.read_csv("f1_qualifying_2000_2026.csv")

drivers = pd.read_csv("f1_driver_standings_2000_2026.csv")

constructors = pd.read_csv("f1_constructor_standings_2000_2026.csv")
```

```python
drivers.head()
```

```python
drivers["performance_score"] = (
    drivers["points"] +
    (drivers["wins"] * 5)
)

drivers[
    ["driver_name", "points", "wins", "performance_score"]
].head()
```

```python
print(drivers.shape)
print(results.shape)
print(qualifying.shape)
```

```python
drivers["performance_score"] = (
    drivers["points"] +
    (drivers["wins"] * 5)
)

top_drivers = drivers.sort_values(
    "performance_score",
    ascending=False
)

top_drivers[
    ["driver_name","points","wins","performance_score"]
].head(10)
```

```python
constructors["dominance_score"] = (
    constructors["points"] +
    (constructors["wins"] * 5)
)

constructors.sort_values(
    "dominance_score",
    ascending=False
)[
    ["constructor","points","wins","dominance_score"]
].head(10)
```

```python
pole_positions = qualifying[
    qualifying["position"] == 1
]

pole_count = (
    pole_positions
    .groupby("driver_name")
    .size()
    .reset_index(name="pole_positions")
    .sort_values(
        "pole_positions",
        ascending=False
    )
)

pole_count.head(10)
```

```python
results["fastest_lap_speed"] = pd.to_numeric(
    results["fastest_lap_speed"],
    errors="coerce"
)

fastest_drivers = (
    results.groupby("driver_name")
    ["fastest_lap_speed"]
    .mean()
    .reset_index()
    .sort_values(
        "fastest_lap_speed",
        ascending=False
    )
)

fastest_drivers.head(10)
```

```python
driver_speed_stats = (
    results.groupby("driver_name")
    .agg(
        avg_speed=("fastest_lap_speed","mean"),
        races=("fastest_lap_speed","count")
    )
)

driver_speed_stats = driver_speed_stats[
    driver_speed_stats["races"] >= 10
]

driver_speed_stats.sort_values(
    "avg_speed",
    ascending=False
)
```
