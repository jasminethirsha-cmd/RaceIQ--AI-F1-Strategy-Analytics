```python
import pandas as pd

races = pd.read_csv("f1_races_2000_2026.csv")
results = pd.read_csv("f1_race_results_2000_2026.csv")
qualifying = pd.read_csv("f1_qualifying_2000_2026.csv")
drivers = pd.read_csv("f1_driver_standings_2000_2026.csv")
constructors = pd.read_csv("f1_constructor_standings_2000_2026.csv")
```

```python
datasets = {
    "races": races,
    "results": results,
    "qualifying": qualifying,
    "drivers": drivers,
    "constructors": constructors
}

for name, df in datasets.items():
    print("\n", "="*50)
    print(name.upper())
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():
    print(name, "duplicates:", df.duplicated().sum())
```

```python
for name, df in datasets.items():
    print("\n" + "="*50)
    print(name.upper())

    missing = df.isnull().sum()

    print(missing[missing > 0])
```
